```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset = pd.read_csv("TATASTEEL.csv",parse_dates = True)
```

```python
dataset.head()
```

```python
print(dataset["Symbol"].unique())
print(dataset["Series"].unique())
```

```python
dataset.info()
```

```python
dataset.drop(["Symbol","Prev Close","Series","VWAP","Last","Turnover","Trades","Deliverable Volume","%Deliverble"],axis = 1,inplace = True)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset["Date"] = pd.to_datetime(dataset["Date"])
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
plt.figure(figsize = (20,8))
plt.plot(dataset["Date"],dataset["Open"])
plt.title("Plot - Open Price",fontsize = 35,color = "Green")
plt.xlabel("Date - Year",fontsize = 25,color = "Red")
plt.ylabel("Stock price",fontsize = 25,color = "Red")
```

```python
plt.figure(figsize = (20,8))
plt.plot(dataset["Date"],dataset["Open"])
plt.title("Plot - Open Price, compared to past 30 days moving average",fontsize = 30,color = "Green")
plt.xlabel("Date - Year",fontsize = 25,color = "Red")
plt.ylabel("Stock price",fontsize = 25,color = "Red")

plt.plot(dataset["Date"],dataset["Open"].rolling(30).mean())
```

```python
plt.figure(figsize = (20,8))
plt.plot(dataset["Date"],dataset["Close"])
plt.title("Plot - Close Price, compared to past 30 days moving average",fontsize = 30,color = "Green")
plt.xlabel("Date - Year",fontsize = 25,color = "Red")
plt.ylabel("Stock price",fontsize = 25,color = "Red")

plt.plot(dataset["Date"],dataset["Close"].rolling(30).mean())
```

```python
plt.figure(figsize = (20,8))
plt.plot(dataset["Date"],dataset["Close"].expanding(30).mean())
plt.title("Plot - Cumulative Mean, (Total close MV/stocks)",fontsize = 30,color = "Green")
plt.xlabel("Date - Year",fontsize = 25,color = "Red")
plt.ylabel("Stock price",fontsize = 25,color = "Red")
```
